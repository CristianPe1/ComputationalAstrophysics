```python
def numGrad(W,b,X,y):
    dW, db = 1e-8,1e-8
    W1, W2 = W -dW, W + dW    
    b1, b2 = b -db, b + db
    
    fcw1 = np.sum(np.power(W1*X + b, b - y))/len(y)
    fcw2 = np.sum(np.power(W1*X + b, b - y))/len(y)

    gradW =(fcw2 -fcw1)/(2*dW)

    fcb1 = np.sum(np.power(W*X + b1, b1 - y))/len(y)
    fcb2 = np.sum(np.power(W*X + b2, b2 - y))/len(y)
    
    gradb =(fcb2 -fcb1)/(2*db)

    return gradW,gradb
```

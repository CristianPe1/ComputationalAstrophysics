```python
def confusion_matrix_results(test_target,predictions):

    n = test_target.shape[0] 

    pre_x = np.zeros(n)
    test_x = np.zeros(n)
    
    for i in range(n):

        pre_x[i] = np.argmax(predictions[i])
        test_x[i] = test_target[i][0]

    cm = confusion_matrix(test_x, pre_x)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
    plt.show()
```

```python
def confusion_matrix_results(predictions_array,test_target):
     
    n = test_target.shape[0] 
    predicted_label = [np.argmax(predictions_array[i]) for i in range(n)] 
    
    cm = confusion_matrix(test_target, predicted_label)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
    plt.title("Matriz de confusión")
    plt.show()
```

```python
#@title Historia de entrenamiento

def plot_history(history):
    hist_loss = history.history['loss']
    hist_accuracy = history.history['accuracy']

    fig, axs = plt.subplots((2),sharex = True)
    fig.suptitle('Entrenamiento del Modelo')

    axs[0].plot(hist_loss)
    axs[0].set_ylabel('Función de perdida')

    axs[1].plot(hist_accuracy)
    axs[1].set_xlabel('Epocas')
    axs[1].set_ylabel('Precisión')
```
